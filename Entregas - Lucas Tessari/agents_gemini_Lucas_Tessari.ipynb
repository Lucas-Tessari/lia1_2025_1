```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
```

```python
# Carregar vari√°vel de ambiente
load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
for model in genai.list_models():
    print(model.name)
```

```python
# Fun√ß√£o para criar agentes simulando comportamentos
def gerar_resposta_gemini(prompt):
    model = genai.GenerativeModel("models/gemini-1.5-flash")
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
# Agente: Planejador de Eventos
def planejar_evento(evento):
    prompt = f"""
    Voc√™ √© um especialista em planejar eventos e garantir uma experi√™ncia inesquec√≠vel para os seus clientes
    Crie um roteiro para o evento {evento}, com margem para 100 pessoas, 200 pessoas e 500 pessoas, incluindo:
    - Local recomendado para o evento com o endere√ßo em Goi√¢nia.
    - Criar um tema
    - Sugest√£o de buffet
    - Sugest√£o de decora√ß√£o
    - Organizar a seguran√ßa do evento
    - Atividades principais
    - Sugest√£o de atra√ß√µes
    """
    return gerar_resposta_gemini(prompt)

```

```python
# Agente: Or√ßamentista
def estimar_orcamento(roteiro):
    prompt = f"""
    Com base neste roteiro de eventos abaixo, fa√ßa uma estimativa de or√ßamento detalhada considerando:
    - Loca√ß√£o do espa√ßo (pre√ßo do im√≥vel)
    - Decora√ß√£o (valor m√©dio do tipo de decora√ß√£o para o evento determinado)
    - Alimenta√ß√£o (custo m√©dio do buffet com base na quatidade de pessoas e no tipo de buffet escolhido)
    - Atividades (entreterimento do evento)
    - Ambienta√ß√¢o (valor m√©dio da atra√ß√£o musical espec√≠fica para o evento determinado)
    - Seguran√ßa (valor m√©dio para ter seguran√ßa total no evento determinado, como ambul√¢ncia e seguran√ßa tercerizada)
    ROTEIRO:
    {roteiro}
    """
    return gerar_resposta_gemini(prompt)
```

```python
# Execu√ß√£o
evento = "Casamento"
roteiro = planejar_evento(evento)
orcamento = estimar_orcamento(roteiro)
```

```python
# Resultados
print("üìç Tipo de evento:\n")
print(roteiro)
print("\nüí∞ Estimativa de Or√ßamento:\n")
print(orcamento)
```
